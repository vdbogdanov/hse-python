```python
def encrypt_simple_substitution_cipher(text, substitution_dict):
    """
    Функция для шифрования текста с использованием шифра простой замены.
    """
    encrypted_text = ""
    for char in text:
        encrypted_text += substitution_dict[char]
    return encrypted_text

def decrypt_simple_substitution_cipher(encrypted_text, substitution_dict):
    """
    Функция для расшифрования текста с использованием шифра простой замены.
    """
    reverse_substitution_dict = {v: k for k, v in substitution_dict.items()}
    decrypted_text = ""
    for char in encrypted_text:
        decrypted_text += reverse_substitution_dict[char]
    return decrypted_text

substitution_dict = {
    'h': 'a',
    'e': 'x',
    'l': 'y',
    'o': 'z'
}

text = "hello"
print("Исходный текст:", text)
encrypted_text = encrypt_simple_substitution_cipher(text, substitution_dict)
print("Зашифрованный текст:", encrypted_text)
decrypted_text = decrypt_simple_substitution_cipher(encrypted_text, substitution_dict)
print("Расшифрованный текст:", decrypted_text)
```

```python
def base_affine_cipher(a, b, m, text, encrypt=True):
    output_text = ""
    for char in text.lower():
        input_index_modify = ord(char) - ord('a')
        if encrypt == True:
            output_index_char = (a * input_index_modify + b) % m
        elif encrypt == False:
            output_index_char = pow(a, -1, m) * (input_index_modify - b) % m
        output_index_modify = output_index_char + ord('a')
        output_text += chr(output_index_modify)
    return output_text

a = 5
b = 8
m = 26

text = "hello"
print("Исходный текст:", text)
encrypted_text = base_affine_cipher(a, b, m, text, encrypt=True)
print("Зашифрованный текст:", encrypted_text)
decrypted_text = base_affine_cipher(a, b, m, encrypted_text, encrypt=False)
print("Расшифрованный текст:", decrypted_text)
```

```python
def calculate_a_b(a1, a2, b1, b2, a_history, b_history, index):
    if index == 0:
        return a1, b1
    elif index == 1:
        return a2, b2
    else:
        a = a_history[index-1] * a_history[index-2]
        b = b_history[index-1] + b_history[index-2]
        return a, b

def recurrent_affine_cipher(a1, a2, b1, b2, m, text, encrypt=True):
    output_text = ""
    a_history, b_history = [], []
    for char, index in zip(text.lower(), range(len(text))):
        a, b = calculate_a_b(a1, a2, b1, b2, a_history, b_history, index)
        input_index_modify = ord(char) - ord('a')
        if encrypt == True:
            output_index_char = (a * input_index_modify + b) % m
        elif encrypt == False:
            output_index_char = pow(a, -1, m) * (input_index_modify - b) % m
        output_index_modify = output_index_char + ord('a')
        output_text += chr(output_index_modify)
        a_history.append(a)
        b_history.append(b)
    return output_text

a1 = 5
a2 = 7
b1 = 8
b2 = 10
m = 26

text = "hello"
print("Исходный текст:", text)
encrypted_text = recurrent_affine_cipher(a1, a2, b1, b2, m, text, encrypt=True)
print("Зашифрованный текст:", encrypted_text)
decrypted_text = recurrent_affine_cipher(a1, a2, b1, b2, m, encrypted_text, encrypt=False)
print("Расшифрованный текст:", decrypted_text)
```
